### **1. Descrição do Problema:**




O gerenciamento do tráfego de veículos em ambientes urbanos representa um grande desafio para os órgãos de trânsito das cidades. Realizar a análise de dados de trânsito a partir de informações de tráfego possibilita um melhor planejamento da utilização das vias públicas, gerenciamento de congestionamentos e auxilia na elaboração de políticas públicas de trânsito.

- **Objetivo:**

Este projeto tem como objetivo realizar uma análise do trânsito de uma localidade a partir de informações do quantitativo e tipos de veículos identificados nas vias públicas.

Além disso, possibilitar a preparação dos dados para uma futura criação de modelo de machie learning que possibilite a previsão do trânsito a partir de informações do quantitativo e tipos de veículos identificados nas vias públicas.

Com essas informações será possível realizar um melhor gerenciamento do trânsito em uma cidade e o planejamento de ações, tais como: restrição de tráfego de caminhões em determinados horários, criação de corredores de ônibus, criação de ciclovias, entre outras ações.


- **Tipo de Problema**

Em função da descrição do problema, o modelo a ser elaborado necessitará fazer a associação entre os atributos das informações de tráfego com uma classe (low, normal, heavy e high) caracterizando um problema de **classificação supervisionada**.


- **Premissas e Hipóteses:**

  - **Premissa**: A quantidade e o tipo de veículo em vias urbanas influenciam no trânsito das vias urbanas.
  - **Hipótese**: A partir de um modelo de machine learning, treinado com dados históricos do quantitativo e tipo de veículos, será possível prever a situação do trânsito nas vias públicas de uma cidade.
      A análise dos dados poderá responder aos seguintes questionamentos:
      - Qual é o tipo de veículo que tende a ter uma maior influência na situação do tráfego das vias urbanas?
      - Quais são os horários que apresentam a maior quantidade de ocorrência da situação de tráfego "pesado" ("Heavy")?
      - Qual é o dia da semana apresenta a maior quantidade de ocorrência da situação de tráfego "pesado" ("Heavy")?

- **Restrições e Condições:**

  - **Restrição**: Os dados utilizados para treinar o modelo foram obtidos do site da Kaggle, e representam informações que foram coletadas em uma única via de veículos em uma cidade.
  - **Condição**: A análise do modelo será realizada apenas com os dados históricos obtidos no dataset disponível no site Kaggle.

  
- **Descrição do dataset:**

O dataset foi obtido a partir do site Kaggle (https://www.kaggle.com/datasets/hasibullahaman/traffic-prediction-dataset) e foi criado a partir de informações que foram obtidas por um modelo de visão computacional que detecta quatro tipos de veículos: carros, bicicletas, ônibus e caminhões.

O dataset possui dados referentes a coleta de informações a cada 15 minutos da quantidade e tipo de veículos em uma via pública, ao longo de 2 (dois) meses. Além disso, disponibiliza uma coluna que indica a situação do tráfego categorizada em quatro classes:  *Low, Normal, Heavy e High*. Possibilitando avaliar o nível de congestionamento e as condições de tráfego em diferentes horários e dias da semana.

Os dados do dataset foram disponibilizados no formato CSV e apresenta as seguintes colunas:

- **Time:** Representa a hora da coleta dos dados de tráfego de veículos, sendo um dado do tipo *string*;

- **Date:** Representa a data (apenas o dia, não tem a informação do mês e ano) da coleta dos dados de tráfego de veículos, sendo um dado do tipo *inteiro*;

- **Day of the week:** Representa o dia da semana que foi realizada a coleta dos dados de tráfego de veículos, sendo um dado do tipo *string*;

- **CarCount:** Representa a quantidade de carros contabilizados desde a última coleta realizada, ou seja, durante o período de 15 minutos, sendo um dado do tipo *inteiro*;

- **BikeCount:** Representa a quantidade de bicicletas contabilizados desde a última coleta realizada, ou seja, durante o período de 15 minutos, sendo um dado do tipo *inteiro*;

- **BusCount:** Representa a quantidade de ônibus contabilizados desde a última coleta realizada, ou seja, durante o período de 15 minutos, sendo um dado do tipo *inteiro*;

- **TruckCount:** Representa a quantidade de caminhões contabilizados desde a última coleta realizada, ou seja, durante o período de 15 minutos, sendo um dado do tipo *inteiro*;

- **Total:** O total de veículos, corresponde a soma dos dados CarCount, BikeCount, BusCount e TruckCount, sendo um dado do tipo *inteiro*;

- **Trafic Situation:** Indica a situação do tráfego no momento da coleta, podendo assumir os seguintes valores:
  - **Low**, tráfego baixo;
  - **Normal**, tráfego normal;
  - **High**, tráfego alto;
  - **Heavy**, tráfego muito alto ("pesado").


Em atualizações futuras do dataset, está prevista a inclusão de dados referentes a velocidade dos veículos e novas rotas, possibilitando uma melhor análise para o gerenciamento de tráfego.


#### **3.3 Cálculo da quantidade média de veículos por tipo coletada na monitoração**



A partir dos dados do dataset, foi possível realizar o cálculo da quantidade média de veículos por tipo que era coletada a cada 15 minutos pelos equipamentos de monitoração, para compor o dataset.


```python
# Calculando a média da quantidade de veículos por tipo
media_carro = transito['Qt_carro'].mean()
media_bicicleta = transito['Qt_bicicleta'].mean()
media_onibus = transito['Qt_Onibus'].mean()
media_caminhao = transito['Qt_caminhao'].mean()

# Agrupando as médias em um DataFrame para realizar a plotagem do gráfico
medias = pd.DataFrame({
    'Tipo veiculo': ['Carro', 'Bicicleta', 'Ônibus', 'Caminhão'],
    'Media': [media_carro, media_bicicleta, media_onibus, media_caminhao]
})

# Representando de forma gráfica a média de veículos por tipo
plt.figure(figsize=(10, 6))
sns.barplot(x='Tipo veiculo', y='Media', data=medias, palette='viridis')

plt.xlabel('Tipo de veículo')
plt.ylabel('Quantidade média de veículos')
plt.title('Quantidade média de veículos por tipo coletado a cada 15 minutos de monitoração')
plt.ylim(0, 100) # Define o limite do eixo Y para notas de 0 a 100

# Adicionar os valores das médias em cima das barras no gráfico
ax = plt.gca() # Utiliza o eixo atual
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points')

plt.show()


```

A partir do gráfico de barras anterior, foi possível identificar que o tipo "**carro**" apresentou uma maior quantidade média de veículos, coletada a cada 15 minutos, na via que estava sendo monitorada.

#### **3.4 Análise da influência do tipo de veículo no trânsito**

A partir da análise entre o tipo de veículo e a situação do tráfego foi possível obter informações do quanto determinado tipo veículo pode influenciar no tráfego de uma via.
Para avaliar essa situação, foi utilizado o gráfico de histograma que possibilita representar a distribuição dos dados em uma determinada situação.

```python
# Criação de subplots para possibilitar a visualização dos histogramas de todos os tipos de veículos em uma única figura
fig, axes = plt.subplots(2, 2, figsize=(15, 10), sharey=True)
fig.suptitle('Análise da quantidade de véiculos por tipo no trânsito')

# O primeiro subplot irá representar a influência de carros no trânsito
sns.histplot(transito,x='Qt_carro',hue='Trafego',kde=True,ax=axes[0,0])
axes[0,0].set_title('Distribuição de Carros')

# O segundo subplot irá reprsentar a influência de bicicletas no trânsito
sns.histplot(transito,x='Qt_bicicleta',hue='Trafego',kde=True,ax=axes[0,1])
axes[0,1].set_title('Distribuição de Bicicletas')

# O terceiro subplot irá representar a influência de ônibus no trânsito
sns.histplot(transito,x='Qt_Onibus',hue='Trafego',kde=True,ax=axes[1,0])
axes[1,0].set_title('Distribuição de Ônibus')

# O quarto subplot irá representar a influência de caminhões no trânsito
sns.histplot(transito,x='Qt_caminhao',hue='Trafego',kde=True,ax=axes[1,1])
axes[1,1].set_title('Distribuição de Caminhões')

plt.subplots_adjust(hspace=0.3)
plt.show()
```

A partir das informações disponibilizadas nos histogramas anteriores foi possível realizar a seguinte análise:

*   **Para ônibus, não é percebido impacto considerável no trânsito**, apresentando alguma influência a partir de aproximadamente 25 ônibus que teria uma indicação de tráfego "*heavy*" (pesado);
*   **Para caminhões, é percebido um impacto mais significativo no trânsito com poucos caminhões na via**, ou seja, com menos de 5 (cinco) caminhões na via já ocorre uma indicação de tráfego "*heavy*" (pesado);

Com o aumento da quantidade de caminhões não é percebida a ocorrência de tráfego "*heavy*" (pesado), uma hipotese para essa situação seria uma preferência dos caminhões em utilizar a via em horários que tenha o tráfego reduzido de outros tipos de veículos. #### **4.2 Conversão de dados para o tipo numérico**



Os modelos de machine learning necessitam que os dados sejam do tipo numérico (*int ou float*) para serem avaliados. O dataset "**transito**" possue os atributos "**Trafego**", "**Dia_semana**" e "**Hora**" que necessitam realizar a conversão para um tipo numérico.

Para os atributos “**Trafego**” e “**Dia_semana**” será realizada a conversão para o tipo numérico fazendo uso da **técnica de mapeamento**. Onde é atribuído um valor numérico manualmente para cada valor do atributo.

```python
#Tratamento de dados, conversão de dados para o tipo numérico (int ou float)

#Conversão dos atributos "Trafego" e "Dia_semana" utilizando a técnica de mapeamento
transito["Trafego"] = transito["Trafego"].map({"low": 0, "normal": 1, "heavy": 2, "high": 3})
transito["Dia_semana"] = transito["Dia_semana"].map({"Sunday": 0, "Monday": 1, "Tuesday": 2, "Wednesday": 3, "Thursday": 4, "Friday": 5, "Saturday": 6 })
```

O atributo **“Hora”** representa a hora do dia que foi realizada a coleta dos dados. Como a coleta é realizada a cada 15 minutos, e um dia possui 1440 minutos ocorreram durante o dia 96 coletas (1440 min / 15 min).

A hora da coleta de dados irá se repetir no dia seguinte, dessa forma, será necessário apenas fazer a conversão para o formato numérico dos 96 horários em um dia que são realizadas as coletas.

Para essa finalidade, foi utilizado o método **OrdinalEncoder()** que realiza o mapeamento de cada dado de uma variável categorica em um valor numérico único do tipo “*int64*”.

A partir do comando “**transito.dtypes()**” foi possível constatar o resultado das transformações dos dados. E o comando “**transito.head()**” possibilitou uma visualização prévia das 5 primeiras linhas do dataset já convertidos para o tipo numérico.

```python
#Tratamento de dados, conversão de dados para o tipo numérico (int ou float)

#Conversão do atributo "Hora", utilizando a técnica "Ordinal Enconder"
ordinal_encoder = OrdinalEncoder()
transito["Hora"] = ordinal_encoder.fit_transform(transito.Hora.values.reshape(-1,1))

#Visualizar uma amostra do Dataset após a transformação dos atributos categoricos em numericos
print("Tipos dos dados dos atributos do Dataset:\n")
print(transito.dtypes)
print("\nAmostra de dados do Dataset após tratamento dos dados:\n")
transito.head()
```

#### **4.6 Preparação de dados para treinamento e teste**


Para realização do processo de modelagem se faz necessário realizar a preparação dos conjuntos de dados que serão utilizados para treinar e testar os algoritmos de machine learning, possibilitando a modelagem desses algoritmos.

Para isso, foi realizada a preparação dos dados para a análise com os modelos. Onde foram divididos os seguintes grupos:


  - **X_train**: Grupo de treino contendo atributos;
  - **y_train** : Grupo de treino contendo a classe;
  - **X_test** : Grupo de teste contendo atributos;
  - **y_test** : Grupo de teste contendo a classe;

Após a divisão foram obtidos 4761 registros para treino e 1191 registros para teste, conforme resultado obtido a partir do método "**shape**" nos grupos que foram criados.

```python
# Preparação dos dados para análise com o modelos

# Separação em bases de treino e teste (holdout)
array = transito.values
X = array[:,0:7]  # Define as colunas do dataset que são atributos
y = array[:,7]  # Define a coluna do datase que é a classe (target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=7) # Realizada a divisão do grupo de dados para treino e teste


# Informações da quantidade de registros após a divisão dos grupos de dados
print("Dimensões do conjunto de treinamento (features):", X_train.shape,"\n")
print("Dimensões do conjunto de treinamento (alvo):", y_train.shape, "\n")
print("Dimensões do conjunto de teste (features):", X_test.shape, "\n")
print("Dimensões do conjunto de teste (alvo):", y_test.shape)
```

### **4.7 Tratamento do desbalanceamento da classe**

Para realizar o tratamento do desbalanceamento da classe foi utilizada a técnica de **undersampling**, que consiste em reduzir o número de amostras da classe majoritária para igualar o tamanho das classes minoritárias. Proporcionando um melhor balanceando para a classe.

Para isso, foi utilizado o método  **“RandomUnderSampler”** que possibilita remover de forma aleatória instâncias da classe majoritária nos dados de treino.

Após o uso do método foram obtidos **2170** registros para treino.

Sendo possível observar através do gráfico que a quantidade de valores atribuídos a situação de tráfego da classe ficou mais balanceada. Ou seja, com valores mais próximos.

```python
#Tratamento do desbalanceamento da classe

rus = RandomUnderSampler(random_state=42, sampling_strategy = 'majority')
X_train_novo, y_train_novo = rus.fit_resample(X_train, y_train)

#Informações da quantidade de registros após o tratamento do desbalenciamento
print("\nDimensões do desbalanceador (features):", X_train_novo.shape, "\n")
print("Dimensões do desbalanceador (alvo):", y_train_novo.shape, "\n")

valores_numericos = [0,1,2,3]
valores_nominais = ["low","normal","heavy","high"]
plt.hist(y_train_novo)
plt.xticks(valores_numericos, valores_nominais)
plt.title('Distribuição da Variável Trafego')
plt.xlabel('Trafego')
plt.ylabel('Contagem')
plt.show()
```

 Enquanto são registrados, em média, 65 carros por coleta para os outros tipos de veículos são registrados em média entre 12 e 18 veículos. Indicando uma distorção nos intervalos de valores, e com um isso a necessidade da normalização para ajustar essa distorção.

A normalização foi aplicada nos dados de treino obtidos após o processo de  *undersampling* (“**X_train_novo**”).

Para possibilitar um comparativo da análise dos dados com e sem normalização foi utilizado o método "**head()**", que visualiza as 5 primeiras linhas de um dataframe, nos dataframes "**transito**" (sem normalização) e "**transito_normalizado**".

```python
# Inicializar o MinMaxScaler
scaler_norm = MinMaxScaler()


# Aprende min e max do X_train_novo que passo pelo processo de undersampling
scaler_norm.fit(X_train_novo)
X_train_normalized = scaler_norm.transform(X_train_novo)

# Usa a média e o desvio padrão aprendidos de X_train
X_test_normalized = scaler_norm.transform(X_test)

# Exibe as primeiras linhas dos dados normalizados (como DataFrame para melhor visualização)
transito_normalizado = pd.DataFrame(X_train_normalized, columns= ['Hora', 'Data', 'Dia_semana', 'Qt_carro', 'Qt_bicicleta', 'Qt_Onibus', 'Qt_caminhao'])

print("\nPrimeiras 5 linhas dos dados originais:")
print(transito.head())


print("\nPrimeiras 5 linhas dos dados normalizados (treino):")
print(transito_normalizado.head())
#print(y_train.head())
```
